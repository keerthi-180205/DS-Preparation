```python
import seaborn as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# scatterplot(num - num) => axes level function
# (hue. style, size) => helps in deriving the added level of information
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex',style = 'time', size = 'size')
```

```python
# relplot => figure level function
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter',hue = 'sex',style = 'time', size = 'size')
```

```python
#line plot => Time series Data => axes level
gap = px.data.gapminder()
temp_df = gap[gap['country']  == 'India']
sns.lineplot(data = temp_df, x = 'year',y = 'lifeExp')
```

```python
sns.relplot(data = temp_df, x = 'year',y = 'lifeExp', kind = 'line')
```

```python
temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Norway'])]
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
# facet plot => only with figure level function(relplot)

sns.relplot(
    data=temp_df,
    x='year',
    y='lifeExp',
    kind='line',
    col='continent'
)
```

```python
#Distribution Plots(univariate analysis)
#histplot
sns.histplot(data = tips, x = 'tip')
```

```python
sns.displot(data = tips, x = 'tip', kind = 'hist', bins = 20)
```

```python
#hue
sns.displot(data = tips, x = 'tip', kind = 'hist', col = 'sex', bins = 20, element='step')
```

```python
# kdeplot
sns.kdeplot(data = tips, x = 'total_bill', hue = 'sex', fill = True)
```

```python
sns.displot(data = tips, x = 'total_bill', hue = 'sex', fill = True, kind = 'kde')
```

```python
#rugplot
sns.kdeplot(data = tips, x = 'tip', hue = 'sex')
sns.rugplot(data = tips, x = 'tip',hue = 'sex')
```

```python
#bivariate analysis
sns.histplot(data = tips, x = 'total_bill', y = 'tip')
sns.kdeplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
#facetplot for catplot
sns.catplot(data = tips, x = 'sex', y = 'total_bill', col = 'day',kind = 'box', hue = 'smoker')
```

```python
#regplot
sns.regplot(data = tips, x = 'total_bill', y ='tip')
```

```python
#lmplot
sns.lmplot(data = tips, x = 'total_bill', y ='tip', hue = 'sex')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

```python
#pairplot
sns.pairplot(data= iris, hue = 'species')
```

```python
#pairdrid
g = sns.PairGrid(data = iris, hue = 'species')
g.map_offdiag(sns.kdeplot)
g.map_diag(sns.boxplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot) 
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species', vars=['sepal_length','petal_length'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot) 
```

```python
#jointplot
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
