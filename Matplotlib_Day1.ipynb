```python
batsman = pd.read_csv('Datasets/Day_23/sharma-kohli.csv')
```

```python
batsman
```

```python
#line plot
plt.plot(batsman['index'], batsman[ 'V Kohli'])
plt.plot(batsman['index'], batsman[ 'RG Sharma'])
```

```python
#title and axis-labels
plt.plot(batsman['index'], batsman[ 'V Kohli'])
plt.plot(batsman['index'], batsman[ 'RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
# linestyle and linewidth
plt.plot(batsman['index'], batsman[ 'V Kohli'], color = 'green', linestyle = 'solid')
plt.plot(batsman['index'], batsman[ 'RG Sharma'],color = 'black', linestyle = 'dashdot')
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
#marker and markersize
plt.plot(batsman['index'], batsman[ 'V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'o', markersize = 8)
plt.plot(batsman['index'], batsman[ 'RG Sharma'],color = 'black', linestyle = 'dashdot', linewidth = 2, marker = 'o', markersize = 8)
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
```

```python
#legend
plt.plot(batsman['index'], batsman[ 'V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'o', markersize = 8,label = 'Virat')
plt.plot(batsman['index'], batsman[ 'RG Sharma'],color = 'black', linestyle = 'dashdot', linewidth = 2, marker = 'o', markersize = 8, label = 'Rohit')
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend()
```

```python
#limiting-axis
price = [10000, 23874, 93842, 12345, 56748,5000000]
year = [2011, 2015, 2019, 2023, 2027, 2031]
plt.plot(year, price)
plt.ylim(0, 100000)
plt.xlim(2015, 2027)
plt.legend()
```

```python
#grid 
plt.plot(batsman['index'], batsman[ 'V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'o', markersize = 8,label = 'Virat')
plt.plot(batsman['index'], batsman[ 'RG Sharma'],color = 'black', linestyle = 'dashdot', linewidth = 2, marker = 'o', markersize = 8, label = 'Rohit')
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend()
plt.grid()
```

```python
#show
plt.plot(batsman['index'], batsman[ 'V Kohli'], color = 'green', linestyle = 'solid', linewidth = 3, marker = 'o', markersize = 8,label = 'Virat')
plt.plot(batsman['index'], batsman[ 'RG Sharma'],color = 'black', linestyle = 'dashdot', linewidth = 2, marker = 'o', markersize = 8, label = 'Rohit')
plt.title('Rohit Sharma vs Virat Kohli IPL Records')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend()
plt.grid()
plt.show()
```

```python
df = pd.read_csv('Datasets/Day_23/batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2, df['2015'],width = 0.2, color= 'blue')
plt.bar(np.arange(df.shape[0]), df['2016'],width = 0.2, color = 'yellow')
plt.bar(np.arange(df.shape[0])+0.2, df['2017'],width = 0.2, color = 'red')
plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
